```python
sns.countplot(x=df['STORECODE'])
plt.title('Count of items purchased from different stores')
plt.show()
```

```python
sns.barplot(x='STORECODE',y='VALUE', hue='MONTH', data=df)
plt.title('Barplot of items purchased at store vs price of item in 3 months')
plt.show()
```

```python
sns.barplot(x='STORECODE',y='QTY', hue='MONTH', data=df)
plt.title('Barplot of items purchased at store vs quantity of item in 3 months')
plt.show()
```

```python
sns.scatterplot(x='QTY',y='VALUE', hue='MONTH', data=df)
plt.title('Scatterplot of Quantity VS Value of item purchased')
plt.show()
```

```python
sns.boxplot(df,vert=True)
plt.show()
```

```python
sns.countplot(x=df['MONTH'],palette='rainbow')
plt.title('Countplot of items purchased in 3 months')
plt.show()
```

```python
sns.pairplot(df,hue='MONTH')
plt.show()
```
